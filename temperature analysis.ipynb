```python
import pandas as pd
import seaborn as sns

sns.set_theme()

"""
Maximum HadCET data:
Daily maximum HadCET, 1878 to date (334 Kb) Format 
https://www.metoffice.gov.uk/hadobs/hadcet/data/download.html
Column 1: year
Column 2: day
Columns 3-14: daily CET values expressed in tenths of a degree. There are 12 columns; one for each of the 12 months.
"""

file_name = "cetdl1772on_max_temperatures_per_daymonth"
data_set = pd.read_csv(file_name, sep="\s+", header=None)

data_set.columns = [
    "year",
    "day",
    "jan",
    "feb",
    "mar",
    "apr",
    "may",
    "jun",
    "jul",
    "aug",
    "sept",
    "oct",
    "nov",
    "dec",
]

```

```python
data_set['jan_celcius'] = data_set.jan/10
data_set['jun_celcius'] = data_set.jun/10
data_set['nov_celcius'] = data_set.nov/10

sub_set= data_set[data_set['day'] ==1]


sns.lmplot(x="year", y="jan_celcius", data=sub_set);

sns.lmplot(x="year", y="jun_celcius", data=sub_set);

sns.lmplot(x="year", y="nov_celcius", data=sub_set);
```

```python
pre_1950 = sub_set[sub_set['year']<= 1950]
post_1950 = sub_set[sub_set['year']> 1950]
pre_industry = sub_set[sub_set['year']< 1860]
post_1970 = sub_set[sub_set['year'] >= 1970]

sns.displot(data=pre_1950, x="jan_celcius",  kind="kde");
sns.displot(data=post_1950, x="jan_celcius",  kind="kde");

```

```python
sns.displot(data=pre_1950, x="jun_celcius",  kind="kde")
sns.displot(data=post_1950, x="jun_celcius",  kind="kde")


```

```python
sns.displot(data=pre_1950, x="nov_celcius", kind="kde");
sns.displot(data=post_1950, x="nov_celcius", kind="kde");
```

```python
sns.regplot(x="year", y="nov_celcius", data=pre_industry)
```
